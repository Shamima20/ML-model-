# 5 Fold Cross-Validation Model Training

## Training different types of model such as Logistic Regression, Random Forest, Gradient Boosting Classifier

```python
from sklearn.model_selection import KFold, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report
```

1 Logistic regression

```python
# Separate features and target
X = dataset.drop("left", axis=1)
y = dataset["left"]

# Define columns
numeric_cols = ["satisfaction_level", "last_evaluation", "number_project",
                "average_montly_hours", "time_spend_company",
                "Work_accident", "promotion_last_5years"]
categorical_cols = ["sales", "salary"]

# transform column
preprocessor = ColumnTransformer([("num", StandardScaler(), numeric_cols),("cat", OneHotEncoder(drop="first"), categorical_cols)])

# create pipeline
pipeline = Pipeline([("preprocessor", preprocessor),("classifier", LogisticRegression(max_iter=1000))])

# 5-Fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
y_pred = cross_val_predict(pipeline, X, y, cv=kf)

# Generate classification report
report = classification_report(y, y_pred, output_dict=True)



report_df = pd.DataFrame(report).T.drop("accuracy", axis=0)

# Plot precision, recall, and f1-score
report_df[["precision", "recall", "f1-score"]].plot(kind='bar', figsize=(8, 5))
plt.title("Logistic Regression - 5-Fold CV Classification Report")
plt.ylabel("Score")
plt.ylim(0, 1.1)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()


print(classification_report(y, y_pred))



```

# Identifying the best model using ROC?AUC and Confusion Matrix

Plot ROC curve and compute AUC

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

models = {
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

```

```python
# Store model ROC data
plt.figure(figsize=(10, 7))
for name, model in models.items():
    y_prob = cross_val_predict(model, X, y, cv=5, method='predict_proba')[:, 1]
    fpr, tpr, _ = roc_curve(y, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")
```

```python
plt.plot([0, 1], [0, 1], 'k--')  # diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

for name, model in models.items():
    y_pred = cross_val_predict(model, X, y, cv=5)
    cm = confusion_matrix(y, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

Which metrics need to be used from confusion matrix? Recall or Precision

```python
# Use Recall for the following scenarios :
Missing a true positive is very costly (e.g., missing an employee who is likely to leave).

You want to catch all actual positives, even at the cost of false positives.

✅ Use Precision if:
False positives are very costly (e.g., wrongly predicting someone will leave could lead to unnecessary HR intervention).

You want accurate positive predictions, even if you miss some real positives.

🔍 In HR Analytics (Employee Attrition Prediction):
Recall is typically more important, because it's more critical to identify all employees at risk of leaving, even if it includes a few who actually wouldn't.

```

```python
# Choosing the Best Model
Compare AUC scores: Higher AUC = better overall classifier

Use confusion matrices to look at recall (TP / [TP + FN])

Pick the model with best balance of recall and acceptable precision
```

# Retention strategies for targeted employees

```python
# Train Random Forest on full data
best_model = RandomForestClassifier(random_state=42)
best_model.fit(X, y)

# Predict probability of leaving (class 1)
turnover_proba = best_model.predict_proba(X)[:, 1]

# Create a DataFrame with probabilities
results_df = X.copy()
results_df['Turnover_Probability'] = turnover_proba
```

```python
# Categorizing into risk zones
def categorize_zone(score):
    if score < 0.2:
        return "Safe Zone (Green)"
    elif score < 0.6:
        return "Low-Risk Zone (Yellow)"
    elif score < 0.85:
        return "High-Risk Zone (Orange)"
    else:
        return "Critical Zone (Red)"
```

```python
results_df['Risk_Zone'] = results_df['Turnover_Probability'].apply(categorize_zone)

# Count of employees per zone
zone_counts = results_df['Risk_Zone'].value_counts().reset_index()
zone_counts.columns = ['Risk_Zone', 'Employee_Count']

zone_counts
```

```python
 Retention Strategies by Zone
 Safe Zone (Green) – Score < 20%
 These employees are highly engaged and satisfied.
Maintain existing policies.
Recognize and reward consistent performers.



Low-Risk Zone (Yellow) – 20% < Score < 60%
 These employees show some signs of disengagement.


Conduct regular check-ins or pulse surveys.

Offer training and development to keep them motivated.

Clarify career progression paths and goals.

High-Risk Zone (Orange) – 60% < Score < 85%
These employees are showing strong signs of possible departure.

Assign mentors or involve them in cross-functional projects.

Provide personalized retention packages.

Resolve workload, recognition, or compensation issues.

Critical Zone (Red) – Score > 85%
These employees are highly likely to leave soon.
```
